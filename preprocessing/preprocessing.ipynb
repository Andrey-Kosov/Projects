since in our sample, the numbers differ in a big way from the state statistics, let's add 3 more categories:

```python
def income_group(total_income):
    if total_income < 20350:
        return '1) до 20350 р.'
    if total_income <= 29300:
        return '2) 20 351 р. - 29 300 р.'
    if total_income <= 40000:
        return '3) 29 301 р. - 40 000 р.'
    if total_income <= 60000:
        return '4) 40 001 р. - 60 000 р.'
    if total_income <= 100000:
        return '5) 60 001 р. - 100 000 р.'
    if total_income <= 175000:
        return '6) 100 001 р. - 175 000 р.'
    if total_income <= 250000:
        return '7) 175 001 р. - 250 000 р.'
    if total_income <= 350000:
        return '8) 250 001 р. - 350 000 р.'
    if total_income <= 450000:
        return '9) 350 001 р - 450 000 р.'
    return '10) более 450 001 р'

data['total_income_group'] = data['total_income'].apply(income_group)
```

```python
def income_group(total_income):
    if total_income < 20350:
        return 'l'
    if total_income <= 29300:
        return '2'
    if total_income <= 40000:
        return '3'
    if total_income <= 60000:
        return '4'
    if total_income <= 100000:
        return '5'
    if total_income <= 175000:
        return '6'
    if total_income <= 250000:
        return '7'
    if total_income <= 350000:
        return '8'
    if total_income <= 450000:
        return '9'
    return '10'

data['total_income_group_id'] = data['total_income'].apply(income_group)
```
