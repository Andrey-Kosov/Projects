# Data description

- gold_recovery_train.csv — train data
- gold_recovery_test.csv — test data
- gold_recovery_full.csv — full data
  
## Technological process

- Rougher feed - raw material
- Rougher additions (or reagent additions) - flotation reagents: Xanthate, Sulphate, Depressant
- Xanthate - xanthate (promoter, or activator of flotation);
- Sulphate - sulfate (sodium sulfide in this production);
- Depressant - depressant (sodium silicate).
- Rougher process - flotation
- Rougher tails - tailings
- Float banks - flotation unit
- Cleaner process - cleaning
- Rougher Au - rough gold concentrate
- Final Au - final gold concentrate



## Stage parameters
- air amount - air volume
- fluid levels - fluid level
- feed size - raw material granule size
- feed rate - feed rate


### Possible values for block [stage]:
- rougher - flotation
- primary_cleaner - primary cleanup
- secondary_cleaner - secondary cleaning
- final - final characteristics

### Possible values for block [parameter_type]:
- input - raw material parameters
- output - product parameters
- state - parameters characterizing the current state of the stage
- calculation - calculated characteristics

## Checking the correctness of the calculation of the enrichment efficiency


- C - the proportion of gold in the concentrate after flotation / refining; *rougher.output.concentrate_au*
- F - the proportion of gold in the raw material / concentrate before flotation / refining; *rougher.input.feed_au*

```python
def recovery(c, f, t):
    return ((c*(f-t))/(f*(c-t)))*100
```

In order to carry out calculations, it is necessary to eliminate the appearance of 0 in the denominator.

```python
len(train_data)
```

```python
train_data = train_data.dropna(
    subset = ['rougher.output.recovery', 'rougher.output.concentrate_au', 
              'rougher.output.tail_au']).reset_index()
len(train_data)
```

```python
train_data = train_data[(train_data['rougher.output.concentrate_au']!=0) & (
    train_data['rougher.output.tail_au']!=0)]
len(train_data)
```

```python
train_data['rougher.output.recovery_check'] = np.where(train_data['rougher.input.feed_au']==0, 0, recovery(
    train_data['rougher.output.concentrate_au'], 
    train_data['rougher.input.feed_au'], 
    train_data['rougher.output.tail_au']))
```

```python
train_data['rougher.output.recovery_check'].describe()
```

```python
train_data['rougher.output.recovery'].describe()
```

```python
print('MAE:', mean_absolute_error(
    train_data['rougher.output.recovery'].dropna(), train_data['rougher.output.recovery_check'].dropna()))
print('MAE: {:.16f}'.format(mean_absolute_error(
    train_data['rougher.output.recovery'].dropna(), train_data['rougher.output.recovery_check'].dropna())))
```

```python
def hist(data, leg):
    plt.hist(data, histtype ='bar', bins=100, alpha=0.5, label=leg)   
```

```python
def hist_start():
    plt.figure(figsize=(14,4)) 

def hist_end(sub):
    plt.suptitle(sub, fontsize=13)
    plt.legend(loc='best')
    plt.show()
```

```python
hist_start()

hist(train_data['rougher.output.recovery'], 'Initial data Recovery')
hist(train_data['rougher.output.recovery_check'], 'Recovery according to our calculations')

hist_end('distribution of enrichment efficiency values')
```

**Conclusions:**
- almost complete coincidence of the distribution of values by histograms (there are two of them: orange and blue)
- MAE: 0.0000000000000097
- The enrichment efficiency is calculated correctly

```python
def concentrate_hist(data, rougher, cleaner, final):
    hist_start()
    hist(data[rougher], rougher)
    hist(data[cleaner], cleaner)
    hist(data[final], final)
```

## A function to calculate the final sMAPE.

```python
def smape(yi_target, yi_predict):
    return 1/len(yi_target) * np.sum(np.abs(yi_predict-yi_target) / ((np.abs(yi_target) + np.abs(yi_predict))/2)*100)
```

```python
def fin_smape(r_smape, f_smape):
    return 0.25 * r_smape + 0.75 * f_smape
```
