# Hypothesis testing - choosing a tariff for a telecom company

## Project description
Customers are offered two tariff plans: "Smart" and "Ultra". In order to adjust the advertising budget, the commercial department wants to understand which tariff brings the most money.  
  
You have to do a preliminary analysis of rates for a small sample of customers. You have at your disposal the data of 500 users: who they are, where they are from, what tariff they use, how many calls and messages each sent in 2018.   
  
It is necessary to analyze the behavior of customers and draw a conclusion - which tariff is better.

## Description of tariffs
###  Tariff "Smart"
Monthly fee: 550 rubles  
Includes 500 minutes of talk time, 50 messages and 15 GB of Internet traffic  
The cost of services in excess of the tariff package:  
- minute of conversation: 3 rubles
- message: 3 rubles
- 1 GB of Internet traffic: 200 rubles
  
### Tariff "Ultra"
Monthly fee: 1950 rubles  
Included 3000 minutes of talk, 1000 messages and 30 GB of Internet traffic  
Cost of services above the tariff package:
- minute of conversation: 1 ruble
- message: 1 ruble
- 1 GB of Internet traffic: 150 rubles

**note:** ISP always rounds up minutes and megabytes. If the user has spoken for only 1 second, a whole minute is counted in the tariff.

# Data description

### Users (information about users):
- user_id - unique user identifier
- first_name - username
- last_name - user's last name
- age - the user's age (years)
- reg_date - tariff activation date (day, month, year)
- churn_date - date of termination of using the tariff (if the value is missing, then the tariff was still in effect - at the time of data upload)
- city - user's city of residence
- tariff - the name of the tariff plan
### Table (information about calls):
- id - unique call number
- call_date - date of the call
- duration - call duration in minutes
- user_id - identifier of the user who made the call
### Messages (information about messages):
- id - unique message number
- message_date - message date
- user_id - identifier of the user who sent the message
### Internet (information about internet sessions):
- id - unique session number
- mb_used - the amount of Internet traffic spent per session (in megabytes)
- session_date - date of the Internet session
- user_id - user ID
### Tariffs :
- tariff_name - tariff name
- rub_monthly_fee - monthly subscription fee in rubles
- minutes_included - the number of minutes of conversation per month included in the subscription fee
- messages_included - number of messages per month included in the subscription fee
- mb_per_month_included - the amount of Internet traffic included in the subscription fee (in megabytes)
- rub_per_minute - the cost of a minute of conversation in excess of the tariff package (for example, if the tariff has 100 minutes of conversation per month, then a fee will be charged from 101 minutes)
- rub_per_message - the cost of sending a message in excess of the tariff package
- rub_per_gb - the cost of an additional gigabyte of Internet traffic in excess of the tariff package (1 gigabyte = 1024 megabytes)

## Converting data to the required types

#### Converting all dates to datetime format

```python
calls['call_date'] = pd.to_datetime(calls['call_date'], format='%Y-%m-%d')
internet['session_date'] = pd.to_datetime(internet['session_date'], format='%Y-%m-%d')
messages['message_date'] = pd.to_datetime(messages['message_date'], format='%Y-%m-%d')
users['churn_date'] = pd.to_datetime(users['churn_date'], format='%Y-%m-%d')
users['reg_date'] = pd.to_datetime(users['reg_date'], format='%Y-%m-%d')
```

#### Round off minutes and megabytes and convert values to int

```python
import math

def ceil(x, step=1): # round up function
    return step * math.ceil(x/step)
```

```python
calls['duration'] = calls['duration'].apply(ceil)
internet['mb_used'] = internet['mb_used'].apply(ceil)
```

```python
calls['duration'] = calls['duration'].astype('int')
internet['mb_used'] = internet['mb_used'].astype('int')
```

## Finding and correcting data errors

- ###  Checking for zero values of call duration

```python
calls_zero = calls.loc[calls['duration'] == 0]
info(calls_zero)
```

Almost 1/5 of all call values are zero. These may be missed calls. calls lasting even 0.1 second are rounded up to a minute. There is no earnings from missed calls and they are not calls. But their recording, along with the calls made, can affect the mean and median. Let's compare these indicators in the dataset without zero calls and with them

```python
calls_no_zero = calls.loc[calls['duration'] > 0].reset_index()

display(calls_no_zero.describe(), calls.describe())
```

**Conclusion:** The presence of records with calls that were not present reduces the average and median call duration by 1.5 - 2 minutes. Such calls have no value, the presence of this data does not add value yet. But just in case, I will leave the original data, and to calculate the averages I will use the calls_no_zero table

- ### **Checking the zero values of Internet traffic**

```python
internet_zero = internet.loc[internet['mb_used'] == 0].reset_index()
info(internet_zero)
internet_no_zero = internet.loc[internet['mb_used'] > 0].reset_index()
```

**Conclusion:** About 15% of internet sessions are recorded as blank. Possibly unsuccessful attempts to connect to the network. Something is possible with data unloading. We will count the number of used dates per user and calculate the average based on the number of users. Our zeros will not interfere with this task. In case we have a problem to calculate the average amount of used data for 1 session, we will resort to the inernet_no_zero table

- ### Monthly revenue from each user

*subtract the free limit from the total number of calls, messages and Internet traffic; multiply the remainder by the value from the tariff plan; add the subscription fee corresponding to the tariff plan*

Create pivot tables with months as the second index value and merge them

```python
calls_no_zero_pivot_merge = calls_no_zero.pivot_table(
    index=('user_id', 'month'), values='duration', aggfunc=('count', 'sum'))

messages_pivot_merge = messages.pivot_table(
    index=('user_id', 'month'), values='id', aggfunc='count')

internet_pivot_merge = internet.pivot_table(
    index=('user_id', 'month'), values='mb_used', aggfunc='sum')
```

Now let's combine all tables into one

```python
data_merge = pd.merge(
    calls_no_zero_pivot_merge, messages_pivot_merge,
    on=['user_id', 'month'], 
    how='outer')
```

```python
data_merge = pd.merge(
    data_merge, internet_pivot_merge,
    on=['user_id', 'month'], 
    how='outer')
```

```python
data_merge.set_axis(
    ['calls_count','calls_minutes','sms_count', 'mb_used'], axis = 'columns', inplace = True)
```

```python
users_merge = users.set_index('user_id')

data_join = data_merge.join(users_merge, on='user_id')
```

```python
tariffs.rename(columns={'tariff_name':'tariff'}, inplace=True)
```

```python
data_join = data_join.join(tariffs.set_index('tariff'), on='tariff')
```

Let's calculate the monthly revenue

(Subtract the free limit from the total number of calls, messages and Internet traffic; multiply the remainder by the value from the tariff plan; add the subscription fee corresponding to the tariff plan).

```python
data_join.head()
```

If there were, for example, calls, but there is no SMS data, then it is reasonable to assume that there was 0 SMS

```python
data_join[['calls_count', 'calls_minutes', 'sms_count', 'mb_used']] = data_join[[
    'calls_count', 'calls_minutes', 'sms_count', 'mb_used']].fillna(0)
```

- Let's calculate the overrun of minutes

```python
data_join['calls_overpay_min'] = data_join['calls_minutes'] - data_join['minutes_included']

data_join.loc[data_join['calls_overpay_min'] < 0, 'calls_overpay_min'] = 0

data_join['calls_overpay_rub'] = data_join['calls_overpay_min'] * data_join['rub_per_minute']
```

- Let's calculate the overrun mb

```python
data_join['overpay_mb'] = data_join['mb_used'] - data_join['mb_per_month_included']

data_join.loc[data_join['overpay_mb'] < 0, 'overpay_mb'] = 0

data_join['overpay_mb_rub'] = (data_join['overpay_mb'] / 1000) * data_join['rub_per_gb']
```

- Calculate SMS overrun

```python
data_join['overpay_sms'] = data_join['sms_count'] - data_join['messages_included']

data_join.loc[data_join['overpay_sms'] < 0, 'overpay_sms'] = 0

data_join['overpay_sms_rub'] = data_join['overpay_sms']  * data_join['rub_per_message']
```

- Let's calculate how much was overpaid in total in excess of the amount of the tariff

```python
data_join['overpay_rub_total'] = data_join['overpay_sms_rub'] + data_join['overpay_mb_rub'] + data_join['calls_overpay_rub']
```

- And how much was paid in the end per month, taking into account the subscription fee:

```python
data_join['paid_per_month'] = data_join['rub_monthly_fee'] + data_join['overpay_rub_total']
```

**Conclusions:**
- All data was summarized in a grouped table by client id and months
- Calculated overrun of minutes, sms and mb
- Cost overruns and total monthly payments calculated
- The table is ready for further analysis
- NaNs are deliberately left unchanged so as not to affect the calculation of averages

# Step 5. Testing hypotheses

Although the answers to these questions have been given before, the indicators were calculated for a sample of 500 users. Now we will judge the general population by the sample taken from it.

We have data on over 20,000 records for 500 users. If the sample was representative (and the distributions seen earlier look normal and tell us that the sample is most likely representative), then its results should most likely characterize the general population.

Accordingly, we will further test the hypothesis about the equality of the means of two general populations

- ### Average revenue of users of Ultra and Smart tariffs differs

   - **Null hypothesis H0:** Average revenue of users of Ultra and Smart tariffs is the same
   - **Hypothesis 1 H1:** Average revenue of users of Ultra and Smart tariffs is statistically significantly different

```python
from scipy import stats as st
```

```python
alpha = .05  #critical level of statistical significance
             # if the p-value is less than it, we reject the hypothesis

results = st.ttest_ind(
    data_ultra['paid_per_month'], 
    data_smart['paid_per_month'], equal_var = False)

print('p-value:', results.pvalue)
print('p-value: {:.70f}'.format(results.pvalue), '...')
print()
if (results.pvalue < alpha):
    print("Rejecting the null hypothesis")
else:
    print("Failed to reject the null hypothesis")
```

**Conclusion:** With tending to 100% * (the probability that the difference happened by chance tends to 0) * with a fraction of probability, we reject the hypothesis that the average revenue of users of two tariffs is the same

- ### The average revenue of users from Moscow differs from the revenue of users from other regions;

```python
data_msc = data_join.loc[data_join['city'] == 'Москва']
data_other = data_join.loc[data_join['city'] != 'Москва']
```

```python
print('Average monthly revenue from subscribers in Moscow:', 
      data_msc['paid_per_month'].mean())
      
print('Average monthly revenue from subscribers in the rest of Russia:', 
      data_other['paid_per_month'].mean())
```

- **Null hypothesis H0:** Average revenue of users from Moscow and other cities is the same
- **Hypothesis 1 H1:** Average revenue of users from Moscow and other cities is statistically significantly different

```python
alpha = .05 

results_city = st.ttest_ind(
    data_msc['paid_per_month'], 
    data_other['paid_per_month']) 

print('p-value:', results_city.pvalue)
print()

if (results_city.pvalue < alpha):
    print("Rejecting the null hypothesis")
else:
    print("Failed to reject the null hypothesis")
```

**Conclusion:** with a high degree of probability (50%), the difference between payments in Moscow and other cities was obtained by chance, it was not possible to reject the null hypothesis that the average revenue in Moscow and other cities is the same
